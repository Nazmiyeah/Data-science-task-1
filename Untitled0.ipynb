```python
import matplotlib.pyplot as plt

# Age groups and population values
age_groups = ['0–20 Years', '21–64 Years', '65+ Years']
population = [512, 807, 98]  # in millions
percentages = [36.1, 57.0, 6.9]
colors = ['yellow', 'dodgerblue', 'hotpink']

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(age_groups, population, color=colors)

# Add data labels on top of each bar
for bar, pop, pct in zip(bars, population, percentages):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 10,
             f'{pop}M\n{pct}%', ha='center', fontsize=10)

# Set Y-axis scale (ticks every 100 million up to 900 million)
plt.yticks(range(0, 901, 100))
plt.ylabel("Population (in millions)")
plt.ylim(0, 900)

plt.xlabel("Age Groups")
plt.xticks(rotation=0)
plt.title("India's Population Distribution by Age (2022)", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
